## 3-1 Activation function 1st : Step Function

```python
def step_function(x):
    if x > 0:
        return 1
    else:
        return 0
```

```python
step_function(-1)
```

```python
step_function(1)
```

```python
import numpy as np
step_function(np.array[-1,1])
```

```python
x = np.array([-1.0, 1.0, 2.0])
x
```

```python
y = x > 0
y
```

```python
y = y.astype(np.int)
y
```

```python
def step_function(x):
    y = x > 0
    return y.astype(np.int)
```

```python
step_function(np.array([-1.0,1.0]))
```

```python
import matplotlib.pylab as plt

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

## 3-2 Activation Function 2nd : Sigmoid Function

```python
def sigmoid(x):
    return 1 / ( 1 + np.exp(-x))
```

```python
x = np.array([-1.0, 1.0, 2.0])
sigmoid(x)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

## 3-3 Activation Function 3rd : Hyperbolic Tangent

```python
def tanh(x):
    return np.tanh(x)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = tanh(x)
plt.plot(x, y)
plt.ylim(-1.1, 1.1)
plt.show()
```

## 3-4 Activation Function 4th : ReLU function

```python
def relu(x):
    if x > 0:
        return x
    else:
        return 0.0
```

```python
relu(1.0)
```

```python
relu(-1.0)
```

```python
x = np.array([-1.0, 1.0, 2.0])
relu(x)
```

```python
def relu(x):
    return np.maximum(x, 0.0)
```

```python
x = np.array([-1.0, 1.0, 2.0])
relu(x)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.ylim(-0.1, 5.1)
plt.show()
```

## 3-5 Activation Function 5th : Linear function 

```python
def linear(x):
    return np.array(x) * 2 + np.ones(1)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = linear(x)
plt.plot(x, y)
plt.ylim(-5.1, 5.1)
plt.show()
```

### 4-1. 1st Step Function : 예 상대적으로 쉬운 유행어(내로남불)를 배우는 것과 같다.
